##Machine Learning Model

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error
```

```python
# Load your dataset
data = pd.read_csv('/content/shipping_cleaned.csv') #drop id column
data.drop(columns=['ID'], inplace=True)
le = LabelEncoder() #encoding categorical columns
for col in data.select_dtypes(include='object').columns:
    data[col] = le.fit_transform(data[col])
X = data.drop('Reached.on.Time_Y.N', axis=1)
y = data['Reached.on.Time_Y.N']
from sklearn.preprocessing import StandardScaler # Scale features for SVM
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
svm_model = SVC(kernel='rbf', random_state=42) #Training model
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test) #predict model
print("SVM Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, y_pred) #confusion matrix
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not On Time', 'On Time'], yticklabels=['Not On Time', 'On Time'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('SVM Confusion Matrix')
plt.show()
```

```python
X = data[['Weight_in_gms']]                      # Only use weight as feature
y = data['Reached.on.Time_Y.N']                 # Target: On time or not
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42) #split data into training values and testing
#Train the SVM model
svm = SVC(kernel='rbf', random_state=42)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Creating the confusion matrix in svm
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, cmap='Purples', fmt='d', xticklabels=['Not On Time', 'On Time'], yticklabels=['Not On Time', 'On Time'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('SVM Confusion Matrix (Weight Only)')
plt.show()
```
